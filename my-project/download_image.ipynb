```python
import pandas as pd
import os
import requests
```

```python
# Load your data
df = pd.read_csv('./public/Food_Stores_Filtered.csv')

# Filter for County
# ny_df = df[df['County'] == 'NEW YORK']
# qn_df = df[(df['County'] == 'QUEENS') & (df['Is Grocery Store'] == 'Yes')]
# br_df = df[(df['County'] == 'BRONX') & (df['Is Grocery Store'] == 'Yes')]
st_df = df[(df['County'] == 'RICHMOND') & (df['Is Grocery Store'] == 'Yes')]



# Extract lat/lon from 'Georeference'
import re
def extract_lat_lon(geo_str):
    match = re.search(r'POINT \((-?\d+\.\d+)\s(-?\d+\.\d+)\)', geo_str)
    if match:
        lon, lat = match.groups()
        return float(lat), float(lon)
    return (None, None)

st_df[['Latitude', 'Longitude']] = st_df['Georeference'].apply(lambda x: pd.Series(extract_lat_lon(x)))

# Create directory to save images
os.makedirs('street_view_images/StatenIsland', exist_ok=True)

# Download each image
for _, row in st_df.iterrows():
    lat, lon = row['Latitude'], row['Longitude']
    license_number = str(row['License Number'])
    url = f"https://maps.googleapis.com/maps/api/streetview?size=640x400&location={lat},{lon}&fov=40&source=outdoor&key=${process.env.GG_API_KEY}"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(f'street_view_images/StatenIsland/{license_number}.jpg', 'wb') as f:
                f.write(response.content)
    except Exception as e:
        print(f"Failed for {license_number}: {e}")
```
